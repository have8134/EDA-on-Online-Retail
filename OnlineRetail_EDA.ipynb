## Plot to find and remove Outliers

Outliers are data points that are very different from the rest of the values in a dataset. They are unusually high or low numbers that don’t follow the normal pattern and can affect averages, charts, and analysis.

```python
plt.boxplot("Quantity", data = df)
```

The plot shows that there are few extreme values beyond the whiskers, indicating the presence of outliers.

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

Q1 and Q3 are the first and third quartiles.

IQR (Interquartile range) measures the middle 50% of the data.

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)
```

Any value outside the Upper & Lower bound range is considered an outlier. They define the range of normal values.

```python
filtered_column = df[(df['Quantity']>=lower_bound) & (df['Quantity']<=upper_bound)]
print(filtered_column)
```

```python
plt.boxplot("Quantity", data = filtered_column)
```

First we removed all the rows where the Quantity value lies outside the IQR boundaries.

After removing outliers, the Boxplot shows more balanced distribution without extreme values, confirming the data has been cleaned successfully.

```python
Q1 = df['UnitPrice'].quantile(0.25)
Q3 = df['UnitPrice'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

## What are the top-selling products by Quantity and Revenue?


```python
df_cleaned = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
print((df_cleaned['Quantity'] <= 0).sum())
```

This step removes all rows with negative quantities or prices, ensuring only valid sales transactions are used for analysis.

```python
top_selling = df_cleaned.groupby("StockCode")["Quantity"].sum()
top_selling = top_selling.sort_values(ascending=False)
print(top_selling)
```

This calculation shows how many units each product has sold in total, allowing us to identify the strongest-selling items based purely on quantity.

```python
top_selling = top_selling.reset_index()
top_15_products_by_quantity = top_selling.head(15)
plt.bar(top_15_products_by_quantity['StockCode'].astype(str), top_15_products_by_quantity['Quantity'])
plt.xlabel('StockCode')
plt.ylabel('Quantity Sold')
plt.title('Top 15 Products Sold')
plt.xticks(rotation=90)
plt.show()
```

The code identifies the top 15 products by total quantity sold and visualizes them using a bar chart to show which stock codes have the highest sales volume. Taller bars indicate items that were purchased most frequently, revealing the strongest volume-driving products.

```python
df_cleaned['TotalRevenue']=df_cleaned['Quantity']*df_cleaned['UnitPrice']
df_cleaned = df_cleaned[~df_cleaned['StockCode'].isin(['POST', 'M'])]
top_seller_Price = df_cleaned.groupby('StockCode')['TotalRevenue'].sum()
top_seller_Price = top_seller_Price.sort_values(ascending=False)
print(top_seller_Price)
```

This step calculates total revenue for each product after removing non-sale stock codes like POST and M, then sorts the results to reveal the highest-earning items. Products at the top of the list generate the most revenue for the business.

```python
top_seller_Price = top_seller_Price.reset_index()
top_15_products_Price = top_seller_Price.head(15)
plt.bar(top_15_products_Price['StockCode'].astype(str), top_15_products_Price['TotalRevenue'])
plt.xlabel('StockCode')
plt.ylabel('Total Revenue')
plt.title('Top 15 Products by Total Revenue')
plt.xticks(rotation=90)
plt.show()
```

The code calculates total revenue for each product and selects the top 15, then plots them in a bar chart to highlight the highest-earning items. Products with taller bars represent stock codes that contribute the most revenue to the business.

## Which Customers bring the Most Revenue (i.e. who are the top buyers)?

```python
df_cleaned['TotalRevenue'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']
top_buyer_Price = df_cleaned.groupby('CustomerID')['TotalRevenue'].sum()
top_buyer_Price = top_buyer_Price.sort_values(ascending=False)
print(top_buyer_Price)
```

This step calculates total revenue contributed by each customer and sorts the results to identify the top buyers. Customers at the top of the list generate the highest overall spending in the dataset.

```python
top_15 = top_buyer_Price.sort_values(by='TotalRevenue', ascending=False).head(15)

plt.bar(top_15['CustomerID'].astype(str), top_15['TotalRevenue'])
plt.xticks(rotation=90)
plt.xlabel('CustomerID')
plt.ylabel('Total Revenue')
plt.title('Top 15 Customers by Revenue')
plt.show()
```

CustomerID 14646 generated the highest revenue in the dataset, followed by 18102 and 17450. These customers account for a large share of total sales and represent the business’s most valuable buyers

## Which items have high sales but low prices — indicating volume drivers?

```python
qty_per_item = df_cleaned.groupby('Description')['Quantity'].sum()
print(qty_per_item)
```

This step calculates how many units of each product were sold in total, helping identify items with high sales volume.

```python
price_per_item = df_cleaned.groupby('Description')['UnitPrice'].mean()
print(price_per_item)
```

This step computes the average price of each product, allowing us to compare item prices and detect low-priced products that may drive volume.

```python
item_data = pd.concat([qty_per_item, price_per_item], axis=1)
item_data.columns = ['TotalQuantity', 'AvgUnitPrice']
print(item_data)
```

This step combines total quantity sold and average unit price into a single table, allowing us to compare both metrics together and identify products that are high-volume and low-priced.

```python
volume_drivers = item_data[
    (item_data['TotalQuantity'] > item_data['TotalQuantity'].quantile(0.75)) & (item_data['AvgUnitPrice'] < item_data['AvgUnitPrice'].quantile(0.25))
    ]

volume_drivers.head()
```

This step filters products to find those with high sales volume and low average prices—items above the 75th percentile in quantity and below the 25th percentile in price. These products represent the key volume drivers in the dataset.

```python
plt.scatter(item_data['AvgUnitPrice'], item_data['TotalQuantity'])
plt.xlabel('Average Unit Price')
plt.ylabel('Total Quantity Sold')
plt.title('Volume Drivers (High Sales, Low Prices)')
plt.show()
```

The scatter plot compares each product’s average unit price with the total quantity sold, helping reveal items that are inexpensive yet sell in very large volumes. Points appearing toward the left (low price) and higher up (high quantity) represent the true volume-driving products in the dataset.

## Which items have low sales but high prices — possible premium products?

```python
qty_per_item = df_cleaned.groupby('Description')['Quantity'].sum()
print(qty_per_item)
```

This step calculates the total units sold for each product, helping identify items with low overall sales volume.

```python
price_per_item = df_cleaned.groupby('Description')['UnitPrice'].mean()
print(price_per_item)
```

This step computes the average price of each product, allowing us to spot items with higher-than-usual pricing that may fall into the premium category.

```python
item_data = pd.concat([qty_per_item, price_per_item], axis=1)
item_data.columns = ['TotalQuantity', 'AvgUnitPrice']
print(item_data)
```

This step merges the total quantity and average price for each product into a single table, making it easier to compare low-volume items with their pricing to identify potential premium products.

```python
premium_product = item_data[
    (item_data['AvgUnitPrice'] > item_data['AvgUnitPrice'].quantile(0.75)) & (item_data['TotalQuantity'] < item_data['TotalQuantity'].quantile(0.25))
    ]

premium_product.head()
```

This step identifies products that have high average prices but very low sales volumes—items above the 75th percentile in price and below the 25th percentile in quantity. These products represent potential premium or niche items in the catalog.

```python
plt.scatter(item_data['TotalQuantity'], item_data['AvgUnitPrice'])
plt.xlabel('Average Unit Price')
plt.ylabel('Total Quantity Sold')
plt.title('Premium Products(Low Sales, High Price)')
plt.show()
```

This plot highlights products that have a high average unit price but sell in low quantities, revealing items that may belong to a premium category. Points appearing far right (high price) but low on the y-axis (low quantity) represent potential premium or niche products.

## Which Countries have highest revenue per customer?

```python
df_cleaned['TotalRevenue'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']
print(df_cleaned['TotalRevenue'])
```

This step calculates the total revenue for each transaction by multiplying quantity and unit price, preparing the dataset for revenue-based analysis.

```python
country_revenue = df_cleaned.groupby('Country')['TotalRevenue'].sum().sort_values(ascending=False)
print(country_revenue)
```

This step aggregates total revenue for each country and sorts the results, revealing which regions contribute the highest sales overall.

```python
top_10_countries = country_revenue.head(10)
print(top_10_countries)
```

This step selects the top 10 revenue-generating countries from the sorted list, making it easy to focus on the strongest-performing markets in terms of total sales.

```python
plt.figure(figsize=(12,6))
plt.bar(top_10_countries.index, top_10_countries.values)
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.title('Top 10 Countries by Total Revenue')
plt.show()
```

The code calculates total revenue for each country and selects the top 10, then plots a bar chart to show which countries contribute the most sales. Countries with taller bars represent regions generating higher revenue in the dataset.
